```python
# We can do the same thing with something like standard deviation, for the sake of brevity I will not perform the "hand calculation"

a = [1,3,5,2,7,8,9,14]  #first we have our array

avg = np.mean(a) #then we need an average to calculate the standard deviation

#now we are ready, first the more cumbersome approach

diffs_sq = [] #for this example, we will need to iterate through values and append a list with some caluclation that we perform as we iterate

#REMEMBER, that the standard deviation is calculated by taking the square root of the quotient between the squared sum of values relative to the mean and the total number of values

#below is a "for loop", which is a method to iterate through data based on a list/array elements
for i in a:  #the notation "i in a" means that the code will consider each element of a sequentially and "i" is the current element in the iteration
  diffs_sq.append((i-avg)**2)  #for each element calculation, we will append the list "diffs_sq", basically filling it in with values
                               #first we calculate the difference between i and avg, then we square that difference "**2" is the notation for squaring a value
print('Squared differences from the mean = '+ str(diffs_sq))

#lastly the take the average of these squarted differences and take the square root

variance = sum(diffs_sq)/len(diffs_sq)  #we can first caluclate the variance

print('Variance = '+str(variance))

standard_dev = variance**.5  ## "**" is an optional notation to denote "to the power of", here to the power of ".5" means "take the square root"

print('Standard Deviation = '+str(standard_dev))
```

```python
#Just like before, we can use in-built functions to simplify our code..
import numpy as np

variance = np.var(a)
standard_dev = np.std(a)

print('Variance = '+str(variance))
print('Standard Deviation = '+str(standard_dev))

#note that we get the same values
```

```python
#the next natural quesiton is to determine if the two groups are statistically different!
#let's first code this manually to see how labor intensive it is...
from scipy.stats import t
control = np.array(df[df['rvirus']=='Control']['%Similarity'])  # here, we take %Sim scores from a df that is filtered to only include "Control"
kd = np.array(df[df['rvirus']=='KD']['%Similarity'])

m1 = np.mean(control)
m2 = np.mean(kd)
sd1 = np.std(control, ddof = 1)
sd2 = np.std(kd, ddof = 1)
pool_s = (sd1**2/len(control) + sd2**2/len(kd))**.5

tt = (m1-m2)/pool_s          #### THIS SHOULD LOOK VERY SIMILAR TO THE FORMULA SHARED IN THE SLIDES!!!!!
dof = len(control)+len(kd)-2
# p-value for 2-sided test
pval = 2*(1 - t.cdf(abs(tt), dof))

print('t-value = '+str(tt)+', pval = '+str(pval))
```

```python
from scipy.stats import linregress  #for the correlation, we will conduct a regression analysis, for which we need a different library

x = df['Accuracy']  #this is how to get column information, based on var name, from a df
y = df['%Similarity']

#x and y are currently lists.. by linregress will expect arrays.. so we need to tranform our lists!

x = np.array(x)
y = np.array(y)

#now we can create a regression model based on our x and y inputs, I will store the model results in the variable "model"

model = linregress(x, y)
model

#notice all of the models' results, e.g. slope, intercept, rvalue, pvalue.....  rvalue is our correlation coefficient! pvalue tells us whether the correlation is significant
#if pvlaue<.05, then significant.. here the pvalue is WELL below .05, therefore there is a significant correlation!

#BUT, WHAT DOES THIS LOOK LIKE IN OUR ABOVE GRAPH?
```

```python
#let's start by plotting the original data
sns.scatterplot(x='Accuracy', y='%Similarity', data=df)
plt.xlim(50, 100)
plt.ylim(0, 100)

#now, let's graph the regression line that describes our correlated graphically.. we need a little math for this (this can aso be done with a
#                                                                     graphing package like seaborn but for now we will do it manually)

slope = model.slope  #model.slope is a method that is used to capture the slope value of the model results
intercept = model.intercept

rvalue = round(model.rvalue, 4)  #let's round these for display purposes
pvalue = round(model.pvalue, 4)


#the plotting function below uses the conventional y = mx+b to derive the regression line based on our model's results
#I included a label arugment for the regression line, such that the r- and p-values are shown on the graph

plt.plot(x, intercept+x*slope, color='k', linewidth=2, label = 'r = '+str(rvalue)+'; p = '+str(pvalue))

plt.legend() #if you want your legend/label to display, you need to pass this function
plt.ylim(0, 105)
```

```python

#BUT.. is this a reasonable correlation? We need to check the dsitribution of "residuals"

#residuals are defined by the difference between a data point and its predicted counterpart  (Yi - Y-hat), where Y-hat is the predicted value relative to Yi

y_hat = []   #empty array, the loop fills it in with the predicted y values for each x value
for i in x:
  y_hat.append(intercept+slope*i)

dy = []   #empty array, the loop fills it in with the residuals, predicted y - observed y
for i in range(0, len(y_hat)):
  dy.append(y_hat[i]-y[i])

```

```python

#here, WE ARE PLOTTING THE RESIDUALS, THE RED DOT IS THE OBSERVED VALUE, THE BLACK LINE DENOTES THE PREDICTED VALUE
# AND THE BLUE LINES DENOTE THE DIFFERENCE/DISTANCE BETWEEN THEM (RESIDUALS)

fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(x, intercept+x*slope, color='k', linewidth=2, label = 'regression: predicted values as a function of x')




ax.scatter(x, y, c='r', label='observed values')
ax.vlines(x,y,y+dy, color='b',alpha=.5, label='residuals: difference between predicted and observed')
plt.title('Regression fit (black) relative to observed elemenets (red); the difference is denoted as a blue line')
plt.legend()
plt.show()

#let's flip this graph by rotating the regressionline as y=0
fig, ax = plt.subplots(figsize=(12, 7))

ax.axhline(0, color='k', linewidth=2, label = 'regression: predicted values as a function of x')


ax.scatter(x, dy, c='r', label='observed values')
ax.vlines(x,0,dy, color='b',alpha=.5, label='residuals: difference between predicted and observed')
plt.title('Rotated regression to illustarte residuals. They DO NOT look uniformly distributed across x')
plt.legend()
fig.suptitle('The residuals should be randomly and uniformly distributed... There should be NO apparent pattern', fontsize=20)
plt.show()

```
